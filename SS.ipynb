```python
import matplotlib.pyplot as plt
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'L:\Program Files\Tesseract-OCR\tesseract.exe'
import pyautogui
import time
from tkinter import *
import cv2
import numpy as np
import statistics
import fct
from PIL import Image

a = True
count = 0
list_champ = []
rep = 1 #negatif = inf

#while a == True :
while count != rep :
    count = count + 1

    font_initial = cv2.imread('map.png')
    img_screen = fct.sct((1585,743),(1901,1062)) #Screen
    img_screen_rgb = fct.process_rgb2bgr(img_screen) #Conversion cv2
    img_screen_font = fct.process_rgb2bgr(img_screen) #Conversion cv2 copy
    img_screen_hsv_red = fct.process_red(img_screen_rgb) #Keep only red
    img_screen_gray = fct.process_hsv2gray(img_screen_hsv_red) #Conversion

    circles = cv2.HoughCircles(img_screen_gray, cv2.HOUGH_GRADIENT, 1, 1, param1=700, param2=10, minRadius=14, maxRadius=15) #detection

    if np.all(circles) != None :
        circles = np.uint16(np.around(circles))

        for i in circles[0, :] :
            xi = i[0]
            yi = i[1]
            radi = i[2]
            champ_detected = img_screen_font[i[1]-18:i[1]+18 , i[0]-18:i[0]+18] #screen tete 38x38 : y,x + BGR
            champ_detected_masked = fct.masked(champ_detected.copy())

            value_list = []

            if list_champ == [] :
                list_champ.append([champ_detected_masked]+[xi]+[yi]+[0]+[radi])
                timer_0 = time.time() #lance timer

            else : #Recherche dans les champs save
                line_counter = -1

                for z in list_champ :
                    line_counter = line_counter + 1
                    value_list.append(fct.matchim(champ_detected_masked,list_champ,line_counter)) #ajoute a la liste la value la moyenne

                max_avg_line = value_list.index(max(value_list)) #renvoi ligne de meilleur value

                if value_list[max_avg_line] > 0.525 :
                    list_champ[max_avg_line][0] = champ_detected_masked #Rewrite screen
                    list_champ[max_avg_line][1] = xi #Rewrite x
                    list_champ[max_avg_line][2] = yi #Rewrite y
                    list_champ[max_avg_line][3] = max_avg_line #Num√©ro fixe d'image
                    list_champ[max_avg_line][4] = radi #Rewrite y
                    globals()['timer_%s' %max_avg_line] = time.time() #Reset timer

                elif len(list_champ) < 5 :
                    list_champ.append([champ_detected_masked]+[xi]+[yi]+[line_counter+1]+[radi])
                    globals()['timer_%s' %(line_counter+1)] = time.time() #Lance timer


    #font = img_screen_font.copy()
    font = font_initial.copy()

    xdelai = -18
    ydelai = -18

    if len(list_champ) > 1 :
        x1 = list_champ[0][1] + xdelai
        y1 = list_champ[0][2] + ydelai
        font[ y1: y1+list_champ[0][0].shape[0] , x1: x1+list_champ[0][0].shape[1] ] = list_champ[0][0]
        globals()['timer_ss_%s' %list_champ[0][3]] = time.time() - globals()['timer_%s' %list_champ[0][3]]
        cv2.putText(font, "{}".format(int(globals()['timer_ss_%s' %list_champ[0][3]])) , (list_champ[0][1]-6,list_champ[0][2]-16), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,255,0), 1, 1)
    if len(list_champ) > 2 :
        x2 = list_champ[1][1] + xdelai
        y2 = list_champ[1][2] + ydelai
        font[ y2: y2+list_champ[1][0].shape[0] , x2: x2+list_champ[1][0].shape[1] ] = list_champ[1][0]
        globals()['timer_ss_%s' %list_champ[1][3]] = time.time() - globals()['timer_%s' %list_champ[1][3]]
        cv2.putText(font, "{}".format(int(globals()['timer_ss_%s' %list_champ[1][3]])) , (list_champ[1][1]-6,list_champ[1][2]-16), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,255,0), 1, 1)
    if len(list_champ) > 3 :
        x3 = list_champ[2][1] + xdelai
        y3 = list_champ[2][2] + ydelai
        font[ y3: y3+list_champ[2][0].shape[0] , x3: x3+list_champ[2][0].shape[1] ] = list_champ[2][0]
        globals()['timer_ss_%s' %list_champ[2][3]] = time.time() - globals()['timer_%s' %list_champ[2][3]]
        cv2.putText(font, "{}".format(int(globals()['timer_ss_%s' %list_champ[1][3]])) , (list_champ[2][1]-6,list_champ[2][2]-16), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,255,0), 1, 1)
    if len(list_champ) > 4 :
        x4 = list_champ[3][1] + xdelai
        y4 = list_champ[3][2] + ydelai
        font[ y4: y4+list_champ[3][0].shape[0] , x4: x4+list_champ[3][0].shape[1] ] = list_champ[3][0]
        globals()['timer_ss_%s' %list_champ[3][3]] = time.time() - globals()['timer_%s' %list_champ[3][3]]
        cv2.putText(font, "{}".format(int(globals()['timer_ss_%s' %list_champ[3][3]])) , (list_champ[3][1]-6,list_champ[3][2]-16), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,255,0), 1, 1)
    if len(list_champ) > 5 :
        x5 = list_champ[4][1] + xdelai
        y5 = list_champ[4][2] + ydelai
        font[ y5: y5+list_champ[4][0].shape[0] , x5: x5+list_champ[4][0].shape[1] ] = list_champ[4][0]
        globals()['timer_ss_%s' %list_champ[4][3]] = time.time() - globals()['timer_%s' %list_champ[4][3]]
        cv2.putText(font, "{}".format(int(globals()['timer_ss_%s' %list_champ[4][3]])) , (list_champ[4][1]-6,list_champ[1][2]-16), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,255,0), 1, 1)


    cv2.imshow("Map",font)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cv2.waitKey()
cv2.destroyAllWindows()
```
